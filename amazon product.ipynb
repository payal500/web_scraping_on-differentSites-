```python
from bs4 import BeautifulSoup as bs
```

```python
import requests
```

```python
url='https://www.amazon.in/OnePlus-Mirror-Black-128GB-Storage/product-reviews/B07DJHV6VZ/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews'
```

```python
page=requests.get(url)
```

```python
page
```

```python
page.content
```

```python
soup=bs(page.text,'html.parser')
```

```python
print(soup.prettify)
```

```python
name=soup.find_all('span', class_='a-profile-name')
```

```python
name
```

```python
name_of_customer=[]
for tags in name:
    name_of_customer.append(tags.text)
name_of_customer
```

```python
name_of_customer.pop(0)
name_of_customer.pop(0)
```

```python
len(name_of_customer)
```

```python
rating=soup.find_all('i',class_="review-rating")
```

```python
rating
```

```python
rating_of_customer=[]
for tags in rating:
    rating_of_customer.append(tags.text)
rating_of_customer.pop(0)
rating_of_customer.pop(0)
len(rating_of_customer)
```

```python
review=soup.find_all('span',{"data-hook":"review-body"})
```

```python
review
```

```python
reviews_of_customer=[]
for tags in range(0,len(review)):
    reviews_of_customer.append(review[tags].get_text())

reviews_of_customer
```

```python
len(reviews_of_customer)
```

```python
reviews_of_customer[:]=[review.rstrip('\n') for review in reviews_of_customer]
```

```python
reviews_of_customer
```
