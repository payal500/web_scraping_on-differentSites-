```python
from bs4 import BeautifulSoup
import requests
url='https://imdb.to/3CzG7te'
r=requests.get(url)
#check the status of page 200 shows sucessfully download 
r.status_code
htmlcontent=r.text
soup=BeautifulSoup(htmlcontent,'html.parser')
print(soup.prettify)
```

```python
topics_title_tags=soup.find_all('div',attrs={'class':'lister-item mode-advanced'})
len(topics_title_tags)
```

```python
topics_title_tags
```

```python
name=[]
for movies in topics_title_tags:
    movies_name=movies.h3.a.text
    name.append(movies_name)
```

```python
name
```

```python
topics_title=soup.find_all('span',{'class':'runtime'})
time=[]
for tags in topics_title:
    time.append(tags.text.replace('min',""))
time
```

```python
import pandas as pd
```

```python
df=pd.DataFrame({"movies":name,"time":time})
df
```

```python
df[:1]
```

```python
df=df[df['time' ] ==df['time'].max()]
```
